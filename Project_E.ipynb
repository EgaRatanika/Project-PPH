## Data cleaning

### Subtask:
Clean the data by handling any missing values, addressing potential data type issues, and ensuring data consistency for analysis.


**Reasoning**:
Identify numeric columns that are currently of object type, remove non-numeric characters from these columns, convert them to numeric types, and then handle missing values by imputing with the median for columns with less than half missing values and dropping columns with more than half missing values.



```python
import numpy as np

# Identify columns that are expected to be numeric but are object type
object_cols = df.select_dtypes(include='object').columns
numeric_cols_to_clean = [col for col in object_cols if col not in ['tahun', 'tax_holiday_status', 'tarif_pph_normal']]

# Remove non-numeric characters and convert to numeric
for col in numeric_cols_to_clean:
    # Check if the column is in the DataFrame before cleaning
    if col in df.columns:
        df[col] = df[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
        # Convert to numeric, coercing errors to NaN
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Handle 'tarif_pph_normal' separately to remove '%' and convert to float
if 'tarif_pph_normal' in df.columns:
    df['tarif_pph_normal'] = df['tarif_pph_normal'].astype(str).str.replace('%', '', regex=False)
    df['tarif_pph_normal'] = pd.to_numeric(df['tarif_pph_normal'], errors='coerce') / 100.0

# Address missing values
# Calculate the percentage of missing values for each column
missing_percentage = df.isnull().sum() / len(df) * 100

# Identify columns to drop (more than 50% missing)
cols_to_drop = missing_percentage[missing_percentage > 50].index

# Drop columns with more than 50% missing values
df.drop(columns=cols_to_drop, inplace=True)
print(f"Dropped columns with more than 50% missing values: {list(cols_to_drop)}")

# Impute missing values in remaining columns with the median
for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype in ['float64', 'int64']:
            median_val = df[col].median()
            df[col].fillna(median_val, inplace=True)
            print(f"Imputed missing values in column '{col}' with the median ({median_val:.2f})")
        elif df[col].dtype == 'object':
             # For object type columns with missing values, fill with mode if appropriate or a placeholder
             mode_val = df[col].mode()[0] if not df[col].mode().empty else 'Missing'
             df[col].fillna(mode_val, inplace=True)
             print(f"Imputed missing values in column '{col}' with the mode ('{mode_val}')")


# Verify data types and check for remaining missing values
display(df.info())
print("\nMissing values per column after cleaning:")
print(df.isnull().sum())
```

## Data visualization

### Subtask:
Create relevant and visually appealing charts (with colors) to represent the data analysis findings. This could include time series plots, bar charts, pie charts, or other appropriate visualizations depending on the data's nature.


**Reasoning**:
Create visualizations to represent the data analysis findings. I will create a time series plot for operational metrics, a bar chart for sales and input PPN, and a bar chart for net PPN.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure 'tahun' is treated as a category or string for plotting if it's not already
df['tahun'] = df['tahun'].astype(str)

# 1. Time series plot for operational metrics
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='tahun', y='pendapatan_operasional', marker='o', label='Pendapatan Operasional', color='skyblue')
sns.lineplot(data=df, x='tahun', y='beban_operasional', marker='o', label='Beban Operasional', color='salmon')
sns.lineplot(data=df, x='tahun', y='laba_kotor_penjualan', marker='o', label='Laba Kotor Penjualan', color='lightgreen')
plt.title('Trend of Operational Metrics Over Years')
plt.xlabel('Tahun')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.show()

# 2. Bar chart for sales and input PPN
fig, ax = plt.subplots(figsize=(12, 6))
bar_width = 0.35
years = df['tahun'].unique()
x = np.arange(len(years))

bars1 = ax.bar(x - bar_width/2, df.groupby('tahun')['pendapatan_penjualan'].mean(), bar_width, label='Pendapatan Penjualan', color='gold')
bars2 = ax.bar(x + bar_width/2, df.groupby('tahun')['ppn_masukan'].mean(), bar_width, label='PPN Masukan', color='mediumpurple')

ax.set_xlabel('Tahun')
ax.set_ylabel('Amount')
ax.set_title('Comparison of Pendapatan Penjualan and PPN Masukan Over Years')
ax.set_xticks(x)
ax.set_xticklabels(years)
ax.legend()

plt.show()

# 3. Bar chart for net PPN
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='tahun', y='ppn_netto', palette='viridis')
plt.title('PPN Netto Over Years')
plt.xlabel('Tahun')
plt.ylabel('PPN Netto Amount')
plt.grid(axis='y')
plt.show()
```
